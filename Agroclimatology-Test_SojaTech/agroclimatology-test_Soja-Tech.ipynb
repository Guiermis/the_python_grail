```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\guilh\OneDrive\Documents\trabaio\agroclimatology.csv')
print(df)
```

```python
df['data'] = pd.to_datetime(df['data'], format='%Y%m%d')
```

```python
data_type = df.dtypes
print(data_type)
```

```python
# dicionário de dados dos meses
month_names = {
    1: 'Janeiro',
    2: 'Fevereiro',
    3: 'Março',
    4: 'Abril',
    5: 'Maio',
    6: 'Junho',
    7: 'Julho',
    8: 'Agosto',
    9: 'Setembro',
    10: 'Outubro',
    11: 'Novembro',
    12: 'Dezembro'
}

#dados que posteriormente serão inseridos pelo usuário
start_year = 2020
end_year = 2020
month_num = 4  

#número do mês para nome do mês
month_name = month_names[month_num]

#filtro da cidade pelo código do ibge
Tibagi = 4127502

#filtro de dados para colocá-los no gráfico
filtered_df = df[(df['data'].dt.year >= start_year) & (df['data'].dt.year <= end_year) & (df['data'].dt.month == month_num) & (df['codigo_ibge'] == Tibagi)]
print(filtered_df)
```

```python
plt.style.use('ggplot')

# Plot Temperature
plt.figure(figsize=(10, 6))
plt.plot(filtered_df['data'], filtered_df['T2M'], color='red')
plt.title(f'Temperature at 2 Meters - Tibagi')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.show()

# Plot precipitation
plt.figure(figsize=(10, 6))
plt.plot(filtered_df['data'], filtered_df['PRECTOTCORR'], color='blue')
plt.title('Precipitation - Tibagi')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.show()

# Plot soil moisture
plt.figure(figsize=(10, 6))
plt.plot(filtered_df['data'], filtered_df['GWETPROF'], color='green')
plt.title('Soil Moisture - Tibagi')
plt.xlabel('Date')
plt.ylabel('Soil Moisture')
plt.show()
```
